## Data Cleaning / Feature Engineering

```python
# After all the analysis to fully understand our data
# we will clean our data
final_data.head()
```

```python
# dealing with missing values
final_data.isnull().sum()
```

```python
final_data.shape
```

```python
# Putting the Time column into a list
times = final_data.Time.value_counts().index.tolist()

# this list comprehension will loop through that list and transform each item into an integer
times = [int(i) for i in times]

# using np.mean, we will get the mean of the list
times_mean = np.mean(times)
```

```python
# Using the mean we obtained in the cell above, we will fill the missing Time with its mean
final_data['Time'] = final_data['Time'].fillna(times_mean)
```

```python
# Putting all numerical columns into the 'numerical' variable and all the objects columns into the 'objects' variable
objects = final_data.select_dtypes(include='object').columns.tolist()
numerical = final_data.select_dtypes(exclude='object').columns.tolist()
```

```python
# columns to drop: 'Div' and 'Date'
# removing them from the list 
objects.remove('Div')
objects.remove('Date')
```

```python
# columns to drop: 'Div' and 'Date'
# removing them from the dataset
drop = ['Div', 'Date']
final_data.drop(drop, axis=1, inplace=True)
```

```python
# This should evaluate to true
len(numerical + objects) == len(final_data.columns.tolist())
```

```python
# No missing values
final_data.isnull().sum()
```

```python
# changing the values of the Time column into integer
time = [int(i) for i in final_data.Time]
final_data.Time = time
```

```python
# making 'objects' list and 'numerical' lists
objects = final_data.select_dtypes(include='object').columns.tolist()
numerical = final_data.select_dtypes(exclude='object').columns.tolist()
```

```python
# Encoding categorical values from the "objects" list
# A model does not understand objects/strings, that is why we encode them to numerical values
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

le.fit(final_data['FTR'].values)
final_data['FTR'] = le.transform(final_data['FTR'].values)

le.fit(final_data['HTR'].values)
final_data['HTR'] = le.transform(final_data['HTR'].values)

le.fit(final_data['HomeTeam'].values)
final_data['HomeTeam'] = le.transform(final_data['HomeTeam'].values)

le.fit(final_data['AwayTeam'].values)
final_data['AwayTeam'] = le.transform(final_data['AwayTeam'].values)
```

```python
# Final time result: D (draw) = 1, H (Home) = 2, A (Away) = 0
# Half time result: D (draw) = 1, H (Home) = 2, A (Away) = 0
```

```python
# All columns are numerical now which mean that the model will accept them as input
final_data.head()
```

```python
# splitting data
# Our target: y
# Our inputs: X

y = final_data['FTR'] # our target is to predict who won the game
X = final_data.drop(['FTR'], axis=1) # our input is everything else
```

```python
# Splitting data into training and testing data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# function that takes the training and testing data. Instantiates an object of type DecisionTreeClassifier,
# trained the model, made predictions and tested the accuracy using the metrics mean squared error and accuracy score,
# respectively.
def evaluate_decision_tree(X_train, X_test, y_train, y_test):
    from sklearn.tree import DecisionTreeClassifier
    from sklearn.metrics import mean_squared_error
    from sklearn.metrics import accuracy_score
    model = DecisionTreeClassifier()
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return mean_squared_error(y_test, preds), accuracy_score(y_test, preds)

evaluate_decision_tree(X_train, X_test, y_train, y_test)
    
```

```python
# function that takes the training and testing data. Instantiates an object of type RandomForestClassifier,
# trained the model, made predictions and tested the accuracy using the metrics mean squared error and accuracy score,
# respectively.
def evaluate_random_forest(X_train, X_test, y_train, y_test):
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.metrics import mean_squared_error
    from sklearn.metrics import accuracy_score
    model = RandomForestClassifier()
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return f" Mean squared error: {mean_squared_error(y_test, preds)}, Accuracy score: {accuracy_score(y_test, preds)}"

evaluate_random_forest(X_train, X_test, y_train, y_test)
```
